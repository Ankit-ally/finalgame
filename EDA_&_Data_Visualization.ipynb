**Spread:** How much the data values are dispersed or scattered.<br><br>
It does NOT tell you direction.<br>
It tells you variability.<br>



*   Range: Difference between Min and Max value  **(Max - min)**
*   Variance: average of squared difference between all datapoint and mean [ **average((xi − mean)²)** ]
*   Standard Deviation: Square root of Vairnace [**√Variance**]

```python
n_cols = 3  # number of columns per row
n_rows = 2

plt.figure(figsize=(18, 5 * n_rows))

for i, col in enumerate(Numerical_column, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.show()
```

| Method          | Use When              | Notes                |
| --------------- | --------------------- | -------------------- |
| Log             | Strong right skew     | Most Reliable        |
| Reflect +Log    | Strong Left skew      | Most Reliable        |
| Sqrt            | Moderate skew         | Less aggressive      |
| Box-Cox         | Positive values only  | Statistical approach |
| Yeo-Johnson     | Zero/negative allowed | More flexible        |
| Winsorize       | Extreme outliers      | Keeps data size      |
| Remove Outliers | Data error            | Risky if real        |


**Question: Do we need to handle Skewness all the times?**

```python
n_cols = 3
n_rows = 2

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
axes = axes.flatten()

for i, col in enumerate(Numerical_column):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')

# Remove extra empty plots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#### **Categorical vs Categorical**

Cross tab

```python
pd.crosstab(df['Region'], df['Payment_Method'])
```

Crosstab with Normalized value

```python
pd.crosstab(df['Region'], df['Payment_Method'], normalize='index') * 100
```

**Stacked Bar Chart**

```python
ct = pd.crosstab(df['Region'], df['Payment_Method'], normalize='index')

ct.plot(kind='bar', stacked=True, figsize=(8,6))
plt.ylabel("Proportion")
plt.show()
```

***Strong Comparison Summary*** <br><br>

| Type       | Best Chart  | What to Observe               |
| ---------- | ----------- | ----------------------------- |
| Num vs Num | Scatter     | Linearity, clusters, outliers |
| Num vs Cat | Boxplot     | Median, spread difference     |
| Cat vs Cat | Stacked bar | Proportion comparison         |
